convert folder of html files to a single text file

```python
def extract_paragraphs_to_file(html_file, output_file):
    # Read the HTML file
    with open(html_file, 'r', encoding='utf-8') as file:
        html_content = file.read()

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'lxml')

    # Find all <p> elements
    paragraphs = soup.find_all('p')

    # Append each paragraph to the output file
    with open(output_file, 'a', encoding='utf-8') as file:
        for paragraph in paragraphs:
            file.write(f"{paragraph.text}\n\n")

def process_folder(folder_path, output_file):
    for filename in os.listdir(folder_path):
        if filename.endswith('.html'):
            html_file_path = os.path.join(folder_path, filename)
            extract_paragraphs_to_file(html_file_path, output_file)
            print(f"Processed {filename}")

process_folder(folder_path, output_file_path)
```

```python
# Open the file in read mode and read lines
with open('combined_output.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Add space at the beginning of each line
lines = [' ' + line for line in lines]

# Open the file in write mode and write the modified lines
with open('combined_output.txt', 'w', encoding='utf-8') as file:
    file.writelines(lines)
```

Webscrape the characters names from the wiki 

```python
url = 'https://coppermind.net/wiki/Category:Rosharans'  # Replace with your new URL

# Fetch the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the list of characters
    character_list = soup.find_all('div', class_='mw-category-group')

    # Open a CSV file to write the characters
    with open('characters.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Character Name'])  # Writing the header

        # Extract character names and write to the CSV
        for group in character_list:
            for link in group.find_all('a'):
                writer.writerow([link.get_text()])
                print(f"Added {link.get_text()} to the CSV")
else:
    print('Failed to retrieve the webpage')
```

```python
import csv
import enum

# Read the CSV file and extract the character names
with open('characters.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header
    character_names = [name.lower() for row in reader for name in row[0].split(',')]

# Create an enumeration of character names
CharacterNames = enum.Enum('CharacterNames', {name: i for i, name in enumerate(character_names, start=1)})
```

```python
import re

with open('combined_output.txt', 'r', encoding='utf-8') as file:
    text = file.read().lower()

    # Sort the names from longest to shortest
    character_names = sorted(character_names, key=len, reverse=True)

    for name in character_names:
        # Convert name to lowercase before searching
        name = name.lower()
        # Check if the name exists in the enumeration
        if name in CharacterNames.__members__:
            # Use regex to replace the names
            text = re.sub(r'\b' + re.escape(name) + r'\b', f"#{CharacterNames[name].value}", text)

# Write the modified text back to the file
with open('combined_output.txt', 'w', encoding='utf-8') as file:
    file.write(text)

print('Done')
```
