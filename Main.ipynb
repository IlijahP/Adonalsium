```python
import os
from bs4 import BeautifulSoup

def extract_paragraphs_to_file(html_file, output_file):
    # Read the HTML file
    with open(html_file, 'r', encoding='utf-8') as file:
        html_content = file.read()

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'lxml')

    # Find all <p> elements
    paragraphs = soup.find_all('p')

    # Append each paragraph to the output file
    with open(output_file, 'a', encoding='utf-8') as file:
        for paragraph in paragraphs:
            file.write(f"{paragraph.text}\n\n")

def process_folder(folder_path, output_folder):
    for root, dirs, files in os.walk(folder_path):
        depth = root[len(folder_path) + len(os.path.sep):].count(os.path.sep)
        if depth < 2:
            # Create a unique output file name for each folder
            folder_name = os.path.basename(root)
            output_file = os.path.join(output_folder, f"{folder_name}_output.txt")

            for file in files:
                if file.endswith('.html'):
                    html_file_path = os.path.join(root, file)
                    extract_paragraphs_to_file(html_file_path, output_file)
                    
            print(f"Done processing folder: {root}")
process_folder(folder_path, output_file_path)
```

```python
import os
import glob

# Define the directory containing the text files
directory = 'extracted texts'

# Define the output file
output_file = 'combined_output.txt'

# Get a list of all text files in the directory
text_files = glob.glob(os.path.join(directory, '*.txt'))

# Open the output file in write mode
with open(output_file, 'w', encoding='utf-8') as outfile:
    # Iterate over each text file
    for text_file in text_files:
        # Open the text file in read mode
        with open(text_file, 'r', encoding='utf-8') as infile:
            # Read the contents of the text file
            contents = infile.read()
            # Write the contents to the output file
            outfile.write(contents)
            # Write a newline character to the output file
            outfile.write('\n')

print('Done combining text files.')
```

```python
urls = [ 'https://coppermind.net/wiki/Category:Nalthians', 'https://coppermind.net/wiki/Category:Taldaini','https://coppermind.net/wiki/Category:Selish','https://coppermind.net/wiki/Category:Eelakin','/','https://coppermind.net/wiki/Category:Threnodite','https://coppermind.net/wiki/Category:Scadrians']  # Replace with your list of URLs

for i, url in enumerate(urls):
    # Fetch the webpage
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the list of characters
        character_list = soup.find_all('div', class_='mw-category-group')

        # Open a unique CSV file to write the characters
        with open(f'characters_{i}.csv', 'w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(['Character Name'])  # Writing the header

            # Extract character names and write to the CSV
            for group in character_list:
                for link in group.find_all('a'):
                    writer.writerow([link.get_text()])
                    print(f"Added {link.get_text()} to the CSV")
    else:
        print(f'Failed to retrieve the webpage: {url}')
```

```python
import re
import csv
import enum

# Read the CSV file and extract the character names
with open('characters.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header
    character_names = [row[0].strip().lower() for row in reader]

# Create an enumeration of character names
CharacterNames = enum.Enum('CharacterNames', {name: i for i, name in enumerate(character_names, start=1)})

# Read the text file
with open('combined_output.txt', 'r', encoding='utf-8') as file:
    text = file.read().lower()

# Replace each character name with its enum value
for character in CharacterNames:
    text = re.sub(r'\b' + re.escape(character.name) + r'\b', '#' + str(character.value), text)

# Write the modified text back to the file
with open('combined_output.txt', 'w', encoding='utf-8') as file:
    file.write(text)

print('Done')
# Create an enumeration of character names
```

```python


# Initialize the adjacency matrix with zeros
matrix = [[0 for _ in range(len(character_names))] for _ in range(len(character_names))]

# Initialize the counter
counter = 0
```

```python
# Split the text into words
words = text.split()

# Iterate over the words
for i in range(len(words)):
    # If the word is a character identifier, look at the 15 words after
    if words[i].startswith('#') and words[i][1:].isdigit():
        word = int(words[i][1:])
        words_list = []
        for j in range(i + 1, min(i + 16, len(words))):
            # If the word is a character identifier, add it to the list
            if words[j].startswith('#') and words[j][1:].isdigit():
                other_word = int(words[j][1:])
                if other_word != word:
                    words_list.append(other_word)
        # Update the adjacency matrix and counter based on the words in the list
        for other_word in words_list:
            matrix[word-1][other_word-1] += 1
            counter += 1  # Increment the counter

print('Done')
print(f'Total connections found: {counter}')
```

```python
def test_character_matrix_correspondence(i, character_names, matrix):
    # Get the i-th character
    character = character_names[i]

    # Get the i-th row of the matrix
    matrix_row = matrix[i]

    # Find the indices of the characters to whom the i-th character is connected
    connected_indices = [index for index, value in enumerate(matrix_row) if value > 0]

    # Find the names of the characters to whom the i-th character is connected
    connected_names = [character_names[index] for index in connected_indices]

    # Find the number of connections for the i-th character
    num_connections = len(connected_indices)

    print(f"The character '{character}' has {num_connections} connections.")

    # Print the names of the characters to whom it is connected and how many times they are connected
    for name, index in zip(connected_names, connected_indices):
        print(f"It is connected to {name} {matrix_row[index]} times.")

# Test the function with an example
for i in range(len(character_names)):
    test_character_matrix_correspondence(i, character_names, matrix)
```

import csv

# File paths
csv_file_path = 'characters.csv'
novel_text_path = 'combined_output.txt'
nodes_csv_path = 'nodes.csv'
edges_csv_path = 'edges.csv'

# Load character names and assign IDs
characters = {}  # {character_name: character_id}
with open(csv_file_path, 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for idx, row in enumerate(reader):
        character_name = row[0].strip()
        characters[character_name] = idx

# Create nodes CSV
with open(nodes_csv_path, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Id", "Label"])  # Header
    for name, id in characters.items():
        writer.writerow([id, name])

# Initialize an empty set to track edges and avoid duplicates
edges_set = set()

# Function to process text and determine edges
def process_text_and_determine_edges(text, characters):
    words = text.split()
    for i, word in enumerate(words):
        if word in characters:
            start_index = max(0, i - 15)
            end_index = min(len(words), i + 16)
            for j in range(start_index, end_index):
                if words[j] in characters and words[j] != word:
                    # Create a tuple representing the edge
                    edge = (characters[word], characters[words[j]])
                    # Add edge to set if not already present to avoid duplicates
                    edges_set.add(edge)

# Process the novel's text to find edges
with open(novel_text_path, 'r', encoding='utf-8') as file:
    novel_text = file.read()
process_text_and_determine_edges(novel_text, characters)

# Create edges CSV
with open(edges_csv_path, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Source", "Target"])  # Header
    for edge in edges_set:
        writer.writerow([edge[0], edge[1]])
